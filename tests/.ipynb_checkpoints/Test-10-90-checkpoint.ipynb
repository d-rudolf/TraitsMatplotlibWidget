```python
# calculate the distance vector
def get_dist_vec(points):
    x = []
    for i in range(0,len(points)):
        diff = points[i]-points[0]
        dist = np.sqrt(diff[0]**2 + diff[1]**2)
        x.append(dist)
    return x
```

Removing data outliers

```python
def norm(data):
    data = np.array(data)
    print 'mean: {0}, std {1}'.format(data.mean(), data.std())
    return (data-data.mean())/data.std()
```

```python
# remove outlier
def get_outlier_index(cluster):
    ''' :param cluster: cluster values (y values)  
        :return: index list of outliers
    '''  
    index_list = []
    cluster_norm = norm(cluster)
    for index, item in enumerate(cluster_norm):
        # item in std
        if abs(item) > 1.5:
            index_list.append(index)
    return index_list       
```

```python
def remove_outliers(cluster, index_list):
    '''
    :param cluster: list of x, list of y values
    :param index_list: indices of values to delete
    :return: cluster without outliers
    '''
    cluster_out = []
    for data in cluster:
        data_out = np.delete(data, index_list)
        cluster_out.append(data_out)
    return cluster_out    
```

```python
def clean_data(cluster, num_it):
    '''
    :param cluster: list of x, list of y values
    :num_it: number of iterations
    :return: cleaned cluster with outliers removed
    '''
    for i in range(0,num_it): 
        index_list = get_outlier_index(cluster[1])
        cluster_clean = remove_outliers(cluster, index_list)
        cluster = cluster_clean
    return cluster
```

```python
# number of iterations
num_it = 3
cluster1 = [cluster1_x, cluster1_y]
cluster2 = [cluster2_x, cluster2_y]
cluster = [cluster1, cluster2]
for elem in cluster:
    data = clean_data(elem, num_it)
    if elem == cluster1:        
        cluster1_clean_x = data[0]
        cluster1_clean_y = data[1]
    else:
        cluster2_clean_x = data[0]
        cluster2_clean_y = data[1]
```

1. Take the mean of cluster 1 and 2 as y_min and y_max. The values for which to extract the x component are y_10 = y_min+0.1*(y_max-y_min) and y_90 = y_min+0.9*(y_max-y_min).
2. Find the zero crossing in e.g. y-y_90, check the deviation from zero, which shouldn't be too large.
3. Get the argument of the y value with the smallest deviation. Interpolate eventually.

```python
y_1 = cluster1_clean_y.mean()
y_std = np.array(cluster1_clean_y).std()
print (y_1)
y_2 = cluster2_clean_y.mean()
print (y_2)
y_diff = y_2-y_1
y_10 = y_1+0.1*y_diff
y_90 = y_1 +0.9*y_diff
print (y_10, y_90)
```

```python
minimum = np.min(np.abs(y_err-y_90))
np.argmin(np.abs(y_err-y_90))
```

```python
def get_x_target(x_err,y_err, y_target, y_std):
    """
    calculates x_10 and x_90
    param x_err: x values
    param y_err: y values
    param y_target: target y value 
    param y_std: y error from clusters
    """
    diff = np.abs(y_err-y_target)
    if np.abs(np.min(diff))<y_std:
        print ('The minimum is: {0:.3f}'.format(np.min(diff)))
        arg = np.argmin(diff)
        print ('The y value is: {0:.3f}'.format(y_err[arg]))
        return x_err[arg]
    else: 
        print('Deviation from y_target too large.')
```

```python
start = int(np.max(cluster1_clean_x))
end = int(np.min(cluster2_clean_x)) 
print ('start index: {0}, end index: {1}'.format(start, end))
x_90 = get_x_target(x_err[start:end], y_err[start:end], y_90, y_std)
x_10 = get_x_target(x_err[start:end], y_err[start:end], y_10, y_std)
print np.abs(x_90-x_10)
```

```python
def find_subsequence(seq, subseq):
    target = np.dot(subseq, subseq)
    candidates = np.where(np.correlate(seq,
                                       subseq, mode='valid') == target)[0]
    # some of the candidates entries may be false positives, double check
    check = candidates[:, np.newaxis] + np.arange(len(subseq))
    mask = np.all((np.take(seq, check) == subseq), axis=-1)
    return candidates[mask]
```

```python
pattern = np.array([1,1,1])
array = np.array([-10,-8,1,1,1,1,2,-9,1,1])
find_subsequence(array, pattern)
```
